```python
import pandas as pd
```

```python
file_path = 'C:/STUDY/busan220608/2day_판다스기초/files/kto_201901.xlsx'
kto_201901 = pd.read_excel(file_path,header=1,skipfooter=4,usecols='A:G')
kto_201901.info()
kto_201901.describe()
```

```python
condition = (kto_201901['관광']==0)|\
            (kto_201901['상용']==0)|\
            (kto_201901['공용']==0)|\
            (kto_201901['유학/연수']==0)
condition
```

```python
print(len(kto_201901[condition]))
kto_201901[condition]
```

```python
# 기준년월 컬럼을 만들고 값은 2019-01로 넣어주세요
kto_201901['기준년월'] = '2019-01'
kto_201901
```

```python
kto_201901['국적'].unique()
```

```python
continents_list=['아시아주', '미주','구주','대양주','아프리카주','기타대륙', '교포소계']
continents_list
```

```python
# 국적 데이터에서 continents_list에 대륙 이름이 없는것만 필터링하기 
```

```python
kto_201901_country = kto_201901[kto_201901['국적'].isin(continents_list)==False]
kto_201901_country
kto_201901_country['국적'].unique()
```

```python
# 기존의 인덱스 삭제하고 인덱스 초기화 하기 
kto_201901_country_newindex = kto_201901_country.reset_index(drop =True)
kto_201901_country_newindex
```

```python
kto_201901[kto_201901['국적'].isin(continents_list)]
```

```python
# 아시아, 아메리카, 유럽 ,오세아니아, 아프리카, 기타대륙, 교포 
continents = ['아시아']*25+['아메리카']*5+['유럽']*23+['오세아니아']*3+['아프리카']*2+['기타대륙']+['교포']
print(len(continents))
continents
```

```python
# 리스트 만들어서 데이터에 넣기 
kto_201901_country_newindex['대륙']=continents
kto_201901_country_newindex
```

```python
kto_201901_country_newindex.head()
```

```python
# 각 국가별 국내에 관광을 위해 입국하는 '관광객비율(%)' 컬럼 생성하기
# 관광객비율(%) = 관광/계*100
```

```python
kto_201901_country_newindex['관광객비율(%)']=round(kto_201901_country_newindex['관광']/kto_201901_country_newindex['계']*100,1)
kto_201901_country_newindex
```

```python
kto_201901_country_newindex.sort_values(by = '관광객비율(%)',ascending=False)
```

```python
# 대륙별 관광객비율 평균 
kto_201901_country_newindex.pivot_table(values = '관광객비율(%)',
                                       index='대륙',
                                       aggfunc='mean')
# index 기준으로 관광객비율이라는 데이터를 mean으로 처리 
```

```python
kto_201901_country_newindex[kto_201901_country_newindex['국적']=='중국']
```

```python
# 컬럼명 : 전체 비율(%) , 소숫점 한자리 
kto_201901_country_newindex['전체비율(%)'] = round(kto_201901_country_newindex['관광']/kto_201901_country_newindex['관광'].sum()*100,1)
kto_201901_country_newindex
```

```python
# '전체비율 구하기 다른식'
round(kto_201901_country_newindex['관광']/sum(kto_201901_country_newindex['관광'])*100,1)
```

```python
# 전체 116개 통합하기
# 통합에 사용할 데이터프레임 변수명 : df
# 전체 통합 후 조회하기
```

```python
i=0
list_name =[]
for i in range(2010, 2020,1):
    for j in range(1,13,1):
        list_name.append((str(i)+str(j).zfill(2)))

list_name
```

```python
clear_df=pd.DataFrame()
for i in list_name:
    try:
        file_path ='C:/STUDY/busan220608/2day_판다스기초/files/kto_{}.xlsx'.format(i) 
        df = pd.read_excel(file_path, header=1,skipfooter=4,usecols='A:G')
        df['기준년월']= i
        df = df[df['국적'].isin(continents_list)==False]
        df=df.reset_index(drop =True)
        df['대륙']=continents
        df['관광객비율(%)']=round(df['관광']/df['계']*100,1)
        df.sort_values(by = '관광객비율(%)',ascending=False)
        df['전체비율(%)'] = round(df['관광']/df['관광'].sum()*100,1)
        clear_df = pd.concat([clear_df,df],ignore_index=True)

    except:
        pass

clear_df
```

```python
df2=pd.DataFrame()

for y in range(2010, 2020,1):
    for m in range(1,13,1):
        try : # 예외처리
            #처리필요
            file_nm = str(y)+str(m).zfill(2)
            file_path = 'C:/STUDY/busan220608/2day_판다스기초/files/kto_{}.xlsx'.format(file_nm)
            kto_201901 = pd.read_excel(file_path,header=1,skipfooter=4,usecols='A:G')

            #처리필요
            kto_201901['기준년월'] = str(y)+'-'+str(m).zfill(2)

            continents_list=['아시아주', '미주','구주','대양주','아프리카주','기타대륙', '교포소계']
            kto_201901[kto_201901['국적'].isin(continents_list)]
            kto_201901_country = kto_201901[kto_201901['국적'].isin(continents_list)==False]
            kto_201901_country_newindex = kto_201901_country.reset_index(drop =True)

            continents = ['아시아']*25+['아메리카']*5+['유럽']*23+['오세아니아']*3+['아프리카']*2+['기타대륙']+['교포']
            print(len(continents))

            kto_201901_country_newindex['대륙']=continents

            kto_201901_country_newindex['관광객비율(%)']=round(kto_201901_country_newindex['관광']/kto_201901_country_newindex['계']*100,1)
            kto_201901_country_newindex.sort_values(by = '관광객비율(%)',ascending=False)
            kto_201901_country_newindex['전체비율(%)'] = round(kto_201901_country_newindex['관광']/kto_201901_country_newindex['관광'].sum()*100,1)

            df2 =  pd.concat([df2,kto_201901_country_newindex],ignore_index=True)
        
        except : 
            pass # 오류나오면 pass 
```
