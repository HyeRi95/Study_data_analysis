- 국적코드표 데이터 읽어들이기

```python
code_master = pd.read_excel('./files3/sample_codemaster.xlsx')
code_master
```

- 원본에 국적명 통합하기 merge()

- merge() 함수 : 왼쪽 데이터 기준으로 오른쪽 데이터가 왼쪽과 같은 값이 있으면 합병하는 함수
         pd.merge(left = 데이터1 ,
             right = 데이터2,
             how = inner/ outer 등등 
             left_on = 데이터1에서 합치고 싶은 컬럼
             right_on = 데이터2에서 합치고 싶은 컬럼 

```python
pd.merge(left = sample_1,
         right = code_master,
         how = 'inner',
         left_on = '국적코드',
         right_on = '국적코드'
         
         )
```
