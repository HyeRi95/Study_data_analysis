```python
code_master2 = pd.read_excel('C:/STUDY/busan220608/1day_파이썬기초/files/sample_codemaster.xlsx')
i=0 # 전역변수 
sample_total = pd.DataFrame() # 전역변수 
for yyyy in range(2019,2021,1):
    for mm in range(1,13,1):
        #print(str(yyyy)+'-'+str(mm).zfill(2))
    
        file_path ='C:/STUDY/busan220608/2day_판다스기초/files2/sample({}).xlsx'.format(i) # 해당 {}안에 i 대입 
        sample2 = pd.read_excel(file_path, header=1, skipfooter=2,usecols = 'A:C')
        i +=1 # 전역변수 i 0에서 시작 +1씩 증가 
        #print(sample2)
        sample2['기준년월'] = str(yyyy)+'-'+str(mm).zfill(2)
        sample_end=pd.merge(left=sample2, 
                            right=code_master2, 
                            how ='left',
                            left_on ='국적코드',
                            right_on='국적코드')
        sample_total = pd.concat([sample_total,sample_end],ignore_index=True) # 누적시켜 주기위해서 전역변수지정하고 다시 저장 
print(len(sample_total))
sample_total
    
```
