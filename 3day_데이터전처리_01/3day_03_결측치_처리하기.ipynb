```python
# 결측치가 있는 모든 데이터를 0으로 변환하기 
df.fillna(0)
```

```python
# 이용거리 컬럼에 있는 결측치만 0으로
df['이용거리'].fillna(0)
```

```python
# 전체 데이터에서 결측치 값을 각 컬럼의 평균값으로 대체하기 
df.fillna(df.mean())
```

```python
# 이용거리 데이터 중에 결측치를 이용거리 평균값으로 대체하기
df['이용거리'].fillna(df['이용거리'].mean())
```
