> 극단치값 계산하기

- Minimum = Q1-1.5*IQR 최소 극단치
- Maximum = Q3+1.5*IQR 최대 극단치
- IQR = Q3-Q1
- Q1(25%)
- Q3(75%)

```python
import numpy as np
```

 Q1(25%), Q3(75%) 데이터 확인하기

```python
Q1, Q3 = np.percentile(df_drop_allow['나이'],[25,75])
print(Q1)
print(Q3)
```

```python
iqr = Q3-Q1
print(iqr)
```

```python
# 최대 극단치 변수 : upper_bound
# 최소 극단치 변수 : lower_bound
lower_bound = Q1-1.5*iqr
upper_bound = Q3+1.5*iqr
print(lower_bound)
print(upper_bound)
```

```python
df_drop_allow[(df_drop_allow['나이']>=upper_bound )| (df_drop_allow['나이']<=lower_bound)]
```

- 극단치 계산하여 극단치범위에서 벗어난 값들 조회해서 출력하는 함수

```python
def limit (df,a):
    Q1, Q3 = np.percentile(df[a],[25,75])
    iqr = Q3-Q1
    lower_bound = Q1-1.5*iqr
    upper_bound = Q3+1.5*iqr
    print(len(df[(df[a]>=upper_bound )| (df[a]<=lower_bound)]))
    print(upper_bound)
    print(lower_bound)
    return(df[(df[a]>=upper_bound )| (df[a]<=lower_bound)])
```

```python
limit (df_drop_allow,'이용시간')
```

### 함수만들기 강사님 풀이 

```python
def outliers_iqr(dframe, data):
    q1,q3=np.percentile(data,[25,75])
    iqr = q3-q1
    upper_bound = q3+(iqr*1.5)
    lower_bound = q1-(iqr*1.5)
    condition = (data >= upper_bound) | (data<=lower_bound)
    print(len(dframe[condition]))
    print(upper_bound)
    print(lower_bound)
    return dframe[condition]
```

```python
temp = outliers_iqr(df_drop_allow, df_drop_allow['이용시간'])
temp
```
