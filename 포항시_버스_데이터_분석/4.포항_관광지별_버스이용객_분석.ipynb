```python
import pandas as pd
```

```python
file_path = 'C:/STUDY/busan220608/8_day_포항시_교통카드_사용내역_분석/trfcard_total.csv'
trfcard_total = pd.read_csv(file_path)
trfcard_total
```

```python
# 필요한 컬럼만 뽑아오기 : 승차데이터
trfcard_total_on=trfcard_total[['승차정류장','승차정류장 GPS X','승차정류장 GPS Y']]
trfcard_total_on
```

```python
# 필요한 컬럼만 뽑아오기 : 하차데이터
trfcard_total_off=trfcard_total[['하차정류장','하차정류장 GPS X','하차정류장 GPS Y']]
trfcard_total_off
```

```python
# 하차 정류장 기준으로 groupby 진행 , count()
trfcard_total_freq = trfcard_total_off.groupby(['하차정류장']).count() 
# 하차 정류장별 '하차정류장GPS X',하차정류장GPS Y' 컬럼이 count 되어서 출력
# '하차정류장GPS X' 또는 '하차정류장GPS Y'열에 있는 값은 해당 하차정류장 열에서 내린 승객수와 동일 (count이기 때문에)
# 어느 하차 정류장에서 사람들이 많이 하차했는지 알기 위해 오름차순으로 정렬
trfcard_total_freq=trfcard_total_freq.sort_values('하차정류장 GPS X',ascending=False)
trfcard_total_freq
```

```python
trfcard_total_freq.to_excel('C:/STUDY/busan220608/8_day_포항시_교통카드_사용내역_분석/정류장별_하차승객수.xlsx')
```

```python
# 상위 20개 추출 
trfcard_total_freq_2 = trfcard_total_freq.head(20)
# 인덱스 reset 
trfcard_total_freq_2=trfcard_total_freq_2.reset_index(drop=False)
# 하차정류장GPS X' 또는 '하차정류장GPS Y' 열의 값은 똑같은 값이므로 '하차정류장GPS Y' 열 삭제 
trfcard_total_freq_2.drop(['하차정류장 GPS Y'],axis=1,inplace=True)
# 하차정류장 GPS X 열의 이름을 하차 승객수로 변환 
trfcard_total_freq_2.rename(columns={'하차정류장 GPS X':'하차승객수'},inplace=True)
trfcard_total_freq_2
```

```python
# 승차정류장별 위도 경도를 추출 하기위해
# 데이터에서 중복 제거한후 승차 정류장의 위도경도 추출 
trfcard_total_on2=trfcard_total_on.drop_duplicates()
trfcard_total_on2
```

```python
# 인덱스 재배열 
trfcard_total_on2.reset_index(drop=True,inplace=True)
trfcard_total_on2
```

```python
# 하차정류장별 위도 경도를 추출 하기위해
# 데이터에서 중복 제거한후 승차 정류장의 위도경도 추출 
trfcard_total_off2=trfcard_total_off.drop_duplicates()
trfcard_total_off2
```

```python
# 인덱스 재배열 
trfcard_total_off2.reset_index(drop=True,inplace=True)
trfcard_total_off2
```

```python
# 위에서 새로 만든 하차 정류장별 위도 경도 데이터프레임을 
# 하차정류장별 승객수 상위 20개를 뽑아 새로 만든 데이터 프레임에 merge
new_df = pd.merge(left = trfcard_total_freq_2,
            right = trfcard_total_off2,
            how = 'inner',
            left_on='하차정류장',
            right_on = '하차정류장')

new_df
```

```python
import folium
import json
```

```python
bus_stop_map = folium.Map(
    # 최초에 보여줄 지도의 중앙점 좌표
    # [위도, 경도]
    location = [36.00568611, 129.3616667],
    
    # 지도 스타일 (openstreetmap : 일반적으로 사용되는 도시형 스타일 지도)
    tiles = "openstreetmap",
    #tiles = "Stamen Terrain",
    #tiles = "cartodbpositron",
    
    
    # 최초 보여줄 지도의 스케일
    zoom_start = 10
)
bus_stop_map
```

```python
# 사람들이 많이 하차하는 top 20 정류장 위치 
for i in range(len(new_df)) :
    lat = new_df.loc[i, "하차정류장 GPS Y"]
    lng = new_df.loc[i, "하차정류장 GPS X"]
    #print(lat, lng)
    
    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        #위치
        location = [lat, lng],
        # 채우기 색상
        fill_color = "red",
        # 채우기 불투명도
        fill_opacity = 1,
        # 태두리 색상
        color = "yellow",
        # 태두리 굵기
        weight = 1,
        # 전체 원의 지름
        radius = 3
    ).add_to(bus_stop_map)

bus_stop_map
```
