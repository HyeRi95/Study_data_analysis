#### 환승객/이용객이 많은 승차정류장 위치

```python
import pandas as pd

# 환승객많음30, 이용객많음30 승차정류장 위치찍기
```

```python
df_subt = pd.read_csv('C:/STUDY/busan220608/8_day_포항시_교통카드_사용내역_분석/trfcard_total.csv')
```

```python
# 이용객많은 정류장 30개 뽑아내고
# 환승객이 많은 정류장 30개 뽑아낸다.
```

```python
df_subt_a = df_subt[['환승여부', '승차정류장', '노선명']]
df_subt_a
```

```python
# 환승여부 Y를 제거한 데이터
df_subt_a1 = df_subt_a.copy()
a = df_subt_a[df_subt_a['환승여부'].isin(['Y'])].index
df_subt_a1.drop(a, inplace=True)

df_subt_a1
```

```python
# groupby한 df와 합치기위해 gps df를 따로 만든다
df_subt_b2 = df_subt[['승차정류장', '승차정류장 GPS X', '승차정류장 GPS Y']]
df_subt_b2 = df_subt_b2.drop_duplicates(['승차정류장'], keep='first')
df_subt_b2
```

```python
# 이용객많은 정류장 top30
a = df_subt_a1.groupby(['승차정류장', '환승여부']).count()
a
aa = a.sort_values(by=['노선명'], ascending=False)
aa = aa.rename(columns = {'노선명':'이용객수'})

transfer_N = pd.merge(aa, df_subt_b2, how='outer', on='승차정류장').head(30)
transfer_N
```

```python
# 환승객이 많은 정류장 top30
b = df_subt_a.groupby(['승차정류장', '환승여부']).count()
bb = b.sort_values(by=['환승여부', '노선명'], ascending=False)
bb = bb.rename(columns = {'노선명':'이용객수'})

transfer_Y = pd.merge(bb, df_subt_b2, how='outer', on='승차정류장')
transfer_Y = transfer_Y.drop_duplicates(['승차정류장'],keep='first')
transfer_Y.reset_index(drop=True, inplace=True)
transfer_Y = transfer_Y.head(30)

transfer_Y
```

```python
# 위 데이터프레임 저장
transfer_N.to_csv('./transfer_N.csv', index=False)
transfer_Y.to_csv('./transfer_Y.csv', index=False)
```

```python
import folium
```

```python
transfer_many_map = folium.Map(
    # 최초에 보여줄 지도의 중앙점 좌표
    # [위도, 경도]
    location = [36.022534, 129.371058],

    # 지도 스타일 (openstreetmap : 일반적으로 사용되는 도시형 스타일 지도)
    tiles = "openstreetmap",
    # tiles = 'Stamen Terrain',
    # tiles = 'cartodbpositron',

    # 최초 보여줄 지도의 스케일
    zoom_start = 11
)
```

```python
for i in range(len(transfer_N)):
    lat = transfer_N.loc[i, '승차정류장 GPS Y']
    lng = transfer_N.loc[i, '승차정류장 GPS X']
    # print(lat, lng)

    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        # 위치
        location = [lat, lng],
        # 채우기 색상
        fill_color = 'green',
        # 불투명도
        fill_opacity = 1,
        # 테두리 색상
        color = 'yellow',
        # 테두리 굵기
        weight = 1,
        # 전체 원의 지름
        radius = 3
    ).add_to(transfer_many_map)
```

```python
for i in range(len(transfer_Y)):
    lat2 = transfer_Y.loc[i, '승차정류장 GPS Y']
    lng2 = transfer_Y.loc[i, '승차정류장 GPS X']
    # print(lat, lng)

    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        # 위치
        location = [lat2, lng2],
        # 채우기 색상
        fill_color = 'blue',
        # 불투명도
        fill_opacity = 1,
        # 테두리 색상
        color = 'yellow',
        # 테두리 굵기
        weight = 1,
        # 전체 원의 지름
        radius = 3
    ).add_to(transfer_many_map)

transfer_many_map
```
